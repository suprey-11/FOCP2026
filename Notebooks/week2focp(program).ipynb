```python
celsius = float (input("Enter a temperature in Celsius: "))

fahrenheit = (celsius * 9/5) + 32

print (f"{celsius} C is equivalent to {fahrenheit}F.")
```

```python
students = int(input("How many students? "))
group_size = int(input("Required group size? "))

groups = students // group_size
left_over = students % group_size

group_word = "group" if groups == 1 else "groups"
student_word = "student" if left_over == 1 else "students"

print(f"There will be {groups} {group_word} with {left_over} {student_word} left over.")
```

```python
print("Sweet Distribution Calculator")
print("-----------------------------")

total_sweets = int(input("How many sweets are in the tub? "))
num_pupils = int(input("How many pupils are present today? "))

sweets_each = total_sweets // num_pupils
leftover = total_sweets % num_pupils

print(f"\nEach pupil gets: {sweets_each} sweets")
print(f"Sweets left over: {leftover}")
```
