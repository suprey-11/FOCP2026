```python
BAD_PASSWORDS = ['password', 'letmein', 'sesame', 'hello', 'justinbieber']

password1 = input("Enter a new password: ")
password2 = input("Enter the password again: ")

if password1 == password2:
    if len(password1) >= 8 & len(password1) <= 12:
        if password1 not in BAD_PASSWORDS:
            print("Password Set")
        else:
            print("Error: Password is too common")
    else:
        print("Error: Password must be 8-12 characters long")
else:
    print("Error: Passwords do not match")
```

```python
BAD_PASSWORDS = ['password', 'letmein', 'sesame', 'hello', 'justinbieber']

while True:
    password1 = input("Enter a new password: ")
    password2 = input("Enter the password again: ")

    if password1 != password2:
        print("Error: Passwords do not match")
        continue

    if len(password1) < 8 or len(password1) > 12:
        print("Error: Password must be 8-12 characters long")
        continue

    if password1 in BAD_PASSWORDS:
        print("Error: Password is too common")
        continue

    print("Password Set")
    break
```

```python
print("Seven Times Table:")
a = int(input("Enter a number: "))
for i in range(0, a + 1):
    result = i * 7
    print(f"{i} x 7 = {result}")
```

```python
number = int(input("Enter the times table number (0-12): "))

if number >= 0 and number <= 12:
    for i in range(0, 13):
        result = i * number
        print(f"{i} x {number} = {result}")
else:
    print("Please enter a number between 0 and 12")
```
