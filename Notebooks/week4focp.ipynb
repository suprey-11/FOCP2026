```python
def count_letters(text):
    uppercase_count = 0
    lowercase_count = 0

    for char in text:
        if char.isupper():
            uppercase_count = uppercase_count + 1
        elif char.islower():
            lowercase_count = lowercase_count + 1

    return uppercase_count, lowercase_count
```

```python
user_text = input("Enter a string: ")

upper, lower = count_letters(user_text)

print("Number of uppercase letters:", upper)
print("Number of lowercase letters:", lower)
```

```python
def remove_last_character(text):

    if len(text) > 1:
        return text[:-1]   
    else:
        return text     
```

```python
user_input = input("Enter a string: ")

result = remove_last_character(user_input)

print("Result:", result)
```

```python
def celsius_to_fahrenheit(c):
    # Formula: F = (C * 9/5) + 32
    f = (c * 9/5) + 32
    return f

def fahrenheit_to_celsius(f):
    c = (f - 32) * 5/9
    return c
```

```python
c_temp = float(input("Enter temperature in Celsius: "))
f_temp = celsius_to_fahrenheit(c_temp)
print(c_temp, "Celsius is", f_temp, "Fahrenheit")

f_temp2 = float(input("Enter temperature in Fahrenheit: "))
c_temp2 = fahrenheit_to_celsius(f_temp2)
print(f_temp2, "Fahrenheit is", c_temp2, "Celsius")
```

```python
import statistics  

temperatures = []

for i in range(6):
    user_input = input(f"Enter temperature {i+1} in Celsius (e.g., 25C): ")
    temp = float(user_input.rstrip('Cc'))
    temperatures.append(temp)

maximum_temp = max(temperatures)
minimum_temp = min(temperatures)
mean_temp = statistics.mean(temperatures)

print(f"Maximum temperature: {maximum_temp}C")
print(f"Minimum temperature: {minimum_temp}C")
print(f"Mean temperature: {mean_temp:.2f}C")
```

```python
import statistics 

temperatures = []

print("Enter temperatures in Celsius (e.g., 25C). Press Enter without typing anything to finish.")

while True:
    user_input = input(f"Enter temperature #{len(temperatures)+1}: ")
    
    if user_input == "":
        break
    
    try:
        temp = float(user_input.rstrip('Cc'))
        temperatures.append(temp)
    except ValueError:
        print("Invalid input! Please enter a number followed by 'C'.")

if len(temperatures) == 0:
    print("No temperatures were entered.")
else:
    maximum_temp = max(temperatures)
    minimum_temp = min(temperatures)
    mean_temp = statistics.mean(temperatures)

    print(f"\nMaximum temperature: {maximum_temp}C")
    print(f"Minimum temperature: {minimum_temp}C")
    print(f"Mean temperature: {mean_temp:.2f}C")
```
