```python
import sys
import requests
```

```python
sys.argv = ['Qno.4.ipynb', 'https://www.google.com']
```

```python
url = sys.argv[1]

try:
    response = requests.get(url, timeout=5)
    if response.status_code == 200:
        print(f"The website {url} is working!")
    else:
        print(f"The website {url} returned status code: {response.status_code}")
except requests.exceptions.RequestException as e:
    print(f"Cannot reach the website {url}. Error: {e}")
```

```python
temperatures = []

print("Enter temperatures in Celsius (e.g., 25C). Press Enter to finish:")

while True:
    user_input = input(f"Enter temperature {len(temperatures)+1}: ")
    if user_input == "":
        break
    try:
        temp = float(user_input.rstrip('Cc'))
        temperatures.append(temp)
    except ValueError:
        print("Invalid input! Please enter a number (e.g., 25C).")
```
