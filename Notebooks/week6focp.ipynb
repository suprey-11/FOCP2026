```python
def to_binary(n):

    if n < 0:
        raise ValueError("Number must be positive")
    return bin(n)[2:] 
```

```python
numbers = [5, 10, 255, 1]

for num in numbers:
    print(f"{num} in binary is {to_binary(num)}")
```

```python
user_input = input("Enter a positive integer: ")

try:
    n = int(user_input)
    print(f"{n} in binary is {to_binary(n)}")
except ValueError:
    print("Invalid input! Please enter a positive integer.")
```

```python
def factors(n):

    if n == 0:
        return [] 
    n = abs(n) 
    result = []
    for i in range(1, n+1):
        if n % i == 0:
            result.append(i)
    return result
```

```python
test_numbers = [6, 12, 15, -10, 0]

for num in test_numbers:
    print(f"Factors of {num}: {factors(num)}")
```

```python
def is_prime(n):
    
    if n <= 1:
        return False  
    if n == 2:
        return True  
    if n % 2 == 0:
        return False 

    for i in range(3, int(n**0.5)+1, 2):
        if n % i == 0:
            return False
    return True
```

```python
test_numbers = [1, 2, 3, 4, 5, 16, 17, 19, 20]

for num in test_numbers:
    print(f"{num} is prime? {is_prime(num)}")
```

```python
user_input = input("Enter an integer to check if it is prime: ")

try:
    n = int(user_input)
    if is_prime(n):
        print(f"{n} is a prime number.")
    else:
        print(f"{n} is not a prime number.")
except ValueError:
    print("Invalid input! Please enter a valid integer.")
```

```python
def simple_encrypt(message):
    
    no_spaces = message.replace(" ", "")  
    reversed_message = no_spaces[::-1]    
    return reversed_message
```

```python
test_messages = [
    "Python programming",
    "Encryption 101",
    "   Leading and trailing spaces   "
]

for msg in test_messages:
    encrypted = simple_encrypt(msg)
    print(f"Original: '{msg}' -> Encrypted: '{encrypted}'")
```

```python
user_message = input("Enter a message to encrypt: ")
encrypted_message = simple_encrypt(user_message)
print(f"Encrypted message: '{encrypted_message}'")
```

```python
import random
import string

def hidden_message_encrypt(message):
    interval = random.randint(2, 20)
    encrypted = ""
    for char in message:
        encrypted += char
        for _ in range(interval-1):
            encrypted += random.choice(string.ascii_lowercase)
    return encrypted, interval

def hidden_message_decrypt(encrypted, interval):
    decrypted = ""
    for i in range(0, len(encrypted), interval):
        decrypted += encrypted[i]
    return decrypted
```

```python
message = "send cheese"
encrypted_message, interval_used = hidden_message_encrypt(message)
print(f"Encrypted message: {encrypted_message}")
print(f"Interval used: {interval_used}")
```

```python
decrypted_message = hidden_message_decrypt(encrypted_message, interval_used)
print(f"Decrypted message: {decrypted_message}")
```
