```python
def unique_letters_sorted(s):

    letters_only = [char.lower() for char in s if char.isalpha()]
    
    unique_sorted = sorted(set(letters_only))
    
    return unique_sorted
```

```python
test_strings = ["cheese", "Python Programming!", "Cat"]

for string in test_strings:
    print(f"Original: '{string}' -> Unique sorted letters: {unique_letters_sorted(string)}")
```

```python
user_input = input("Enter a string: ")
print("Unique sorted letters:", unique_letters_sorted(user_input))
```

```python
def letters(word):
    return set(char.lower() for char in word if char.isalpha())
```

```python
def letters_union(word1, word2):
    return sorted(letters(word1) | letters(word2)) 
```

```python
def letters_intersection(word1, word2):
    return sorted(letters(word1) & letters(word2))
```

```python
def letters_symmetric_difference(word1, word2):
    return sorted(letters(word1) ^ letters(word2)) 
```

```python
word1 = input("Enter the first word: ")
word2 = input("Enter the second word: ")

print("\nLetters in at least one of the words (Union):", letters_union(word1, word2))
print("Letters in both words (Intersection):", letters_intersection(word1, word2))
print("Letters in either word but not both (Symmetric Difference):", letters_symmetric_difference(word1, word2))
```

```python
countries = {}
```

```python
while True:
    country = input("Enter a country name (or press Enter to quit): ").strip()

    if country == "":
        print("Program ended.")
        break

    country_key = country.lower()

    if country_key in countries:
        print(f"The capital of {country.title()} is {countries[country_key]}")
    else:
        capital = input(f"I don't know the capital of {country.title()}. Please enter it: ").strip()
        countries[country_key] = capital
        print(f"Saved: {country.title()} → {capital}")
```

```python
print("Countries and capitals stored:")
for country, capital in countries.items():
    print(country.title(), ":", capital)
```

```python
letter_counts = {}

for ch in "abcdefghijklmnopqrstuvwxyz":
    letter_counts[ch] = 0
```

```python
message = input("Enter the encrypted message: ")

for char in message.lower():
    if char.isalpha():
        letter_counts[char] += 1
```

```python
sorted_letters = sorted(letter_counts.items(), key=lambda x: x[1], reverse=True)

print("Six most common letters:")
for letter, count in sorted_letters[:6]:
    print(f"{letter} : {count}")
```
