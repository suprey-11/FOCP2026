```python
REQUIRED_PIZZAS = 4
CURRENCY = "Â£"

def get_price(pizza_number):
    while True:
        user_input = input(f"Enter Price of Pizza #{pizza_number} ({CURRENCY}): ")
        try:
            price = float(user_input)
            if price <= 0:
                print(" > Error: Price must be greater than 0.")
            else:
                return price
        except ValueError:
            print(" > Error: Please enter a numeric value (e.g. 12.50).")

print("---------------------------------")
print(" Beckett Pizza Plaza | 4-for-3 ")
print("---------------------------------\n")

prices = []

for i in range(1, REQUIRED_PIZZAS + 1):
    prices.append(get_price(i))

cheapest = min(prices)
total_before = sum(prices)
total_after = total_before - cheapest

discount_percentage = (cheapest / total_before) * 100

print("\n" + "="*33)
print(f"{'ORDER SUMMARY':^33}")
print("="*33)

tracker_cheapest = cheapest

for i, price in enumerate(prices, 1):
    label = f"Pizza #{i}"
    if price == tracker_cheapest:
        print(f"{label:<15} {CURRENCY}{price:>10.2f} (FREE)")
        tracker_cheapest = -1
    else:
        print(f"{label:<15} {CURRENCY}{price:>10.2f}")

print("-" * 33)
print(f"{'Subtotal:':<15} {CURRENCY}{total_before:>10.2f}")
print(f"{'Discount:':<15}-{CURRENCY}{cheapest:>10.2f}")
print("=" * 33)
print(f"{'TOTAL TO PAY:':<15} {CURRENCY}{total_after:>10.2f}")
print("=" * 33)

print(f"\nYou saved {discount_percentage:.0f}% on this order!")
```
