```python
import random
import sys

print("Enter your password:", end=" ")
password = input()

if len(password) < 9:
    print("\nPassword too short.")
    quit()

positions = []
for _ in range(3):
    index = random.randint(1, len(password))
    positions.append(index)

for pos in positions:
    print(f"\nEnter letter at position {pos}:", end=" ")
    user_letter = input()

    if user_letter == password[pos - 1]:
        print("Correct")
    else:
        print("Security check failed.")
        sys.exit()

print("\nSecurity check passed.")
```
